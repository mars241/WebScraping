### **Scrapons les [cucurbitacées](https://www.graines-semences.com/27-cucurbitacees) sur le site de Graines Semences**

Objectif : construire un DataFrame avec les noms des fruits + leurs prix

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
# Requête et parsing html
url = "https://www.graines-semences.com/27-cucurbitacees"
navigator = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1)'
page = requests.get(url, headers={'User-Agent': navigator})
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

### **Afficher tous les produits, puis isoler le premier**

```python
soup.find_all("div", class_="product-infos")
```

```python
# Afficher tous les produits
# Attention triche : limit 36 sinon on attrappe aussi les actualités en fin de page !
products_soup = soup.select("div.product-infos", limit=36)
products_soup
```

```python
len(products_soup)
```

```python
# Isoler le premier produit
product = products_soup[0]
product
```

### **Nom du premier produit**

```python
# Extraire title ou text
product.select("h3 a")
```

### **Automatisation pour tous les produits**

```python
products_soup
```

```python
product.select("h3 a")[0]['title']
```

```python
cucurbitacees_dict = {'fruit' : [],
                      'price' : []}

# Itérer sur tous les produits trouvés
# appliquer à chacun les fonctions établies auparavant
# et les stocker au fur et à mesure dans un dictionnaire

for product in products_soup:
  fruit = product.select("h3 a")[0]['title']
  price = product.select("span.price")[0].text
  price = re.sub('[\\t\\n]+', '', price)

  cucurbitacees_dict['fruit'].append(fruit)
  cucurbitacees_dict['price'].append(price)

cucurbitacees_dict
```

### **Transformation en DataFrame**

```python
cucurbitacees_dict
```

```python
print(len(cucurbitacees_dict['fruit']))
print(len(cucurbitacees_dict['price']))
```

```python
df_cucurbitacees = pd.DataFrame(cucurbitacees_dict)
df_cucurbitacees
```
