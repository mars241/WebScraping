# 1. Lance une requête sur la page web indiquée ci-dessus.


```python
# importer les bibliothèques pour le WebScrapping/Parser un site
import requests
from bs4 import BeautifulSoup

# Caratéristiques et lien de la page web,puis lancement du "requests"
url = "http://www.chucknorrisfacts.fr/facts/top/1"
html = requests.get(url)
```

```python
# Afficher/accèder à la page Web ci-dessus(url)

page_web = html.text
page_web
```

```python
# Création du "soup" via BeautfulSoup et parsing: c'est-à-dire du nettoyage et de la mise en ordre de ce code HTML

soup = BeautifulSoup(page_web, 'html.parser')
```

```python
# Conterner="Bypass" est la protection en spécifiant un navigateur existant. Et réessayez votre demande!

items_container = soup.find_all("div", {"class", "card"})
items_container
```

# 5. Fais de même pour isoler uniquement la note de la 8e blague

```python
# Affiche le 8ème bloc de blagues en utilisant l'attribut .text.
# NB: le compteur sur python débute par 0

items_container[7].p.text
```

```python
# Utiliser la méthode .find avec la balise appropriée pour isoler uniquement la 8ème blague, sans les autres éléments

items_jokes = soup.find_all(class_= "card-text")
items_jokes[7].text
```

```python
# Faire de même pour isoler uniquement le résultat de la notation de la 8ème blague.

items_rating = soup.find_all("span")
items_rating[7].text
```

# 6. Crée un dictionnaire vide, puis crée une boucle qui viendra enregistrer dans ce dictionnaire chaque blague en clé et chaque note correspondante en valeur.

```python
# Créer un dictionnaire vide, puis créer une boucle qui enregistrera dans ce dictionnaire,
# chaque blague en tant que clé et chaque résultat de notation correspondant en tant que valeur.

# initialisation du dictionaire nommé "joke_rating"
joke_rating ={}

for i in range(len(items_container)): # parcourir les items des conteneures (en anglais"container")
  joke_rating[items_container[i].p.text] = items_container[i].span.text # éliminer les balises <Span> et <script> puis d'autres balises HTML.
  #Autres méthodes de supression de balises: "get_text()"/ "span.unwrap()"/ "span.decompose()" ect..
print(joke_rating)
```

# 7. Transforme ce dictionnaire en DataFrame de 2 colonnes : blague et note. Il doit avoir 20 lignes : une par blague.

```python
# Transforme ce dictionnaire en DataFrame de 2 colonnes : blague et note. Il doit avoir 20 lignes : une par blague.

# import des bibliothèques de lecture (pandas) et stockage (pandas.io.parsers.readers) de DataFram:
from pandas.io.parsers.readers import is_index_col
import pandas as pd

df = pd.DataFrame(joke_rating.items(), columns=['Jokes', 'Rating_Result'])
df
```
